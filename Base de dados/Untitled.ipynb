### Normalizando colunas
Como as colunas estão com os dados em formato json, precisamos normalizar para extrair os dados

```python
customer_df = pd.json_normalize(df.customer)
customer_df
```

```python
phone_df = pd.json_normalize(df.phone)
phone_df
```

```python
internet_df = pd.json_normalize(df.internet)
internet_df
```

```python
account_df = pd.json_normalize(df.account)
account_df
```

```python
df
```

### Unindo DataFrames gerados

Os DF gerados após a normalização das colunas precisam ser unidos para que possamos 
dar continuidade no processo de análise dos dados. 
Vamos unir pelo índice e de forma lateral, já que os dados tem os mesmos índices

```python
dados = pd.merge(
    left= customer_df,
    right= phone_df,
    left_index=True,
    right_index= True
)
```

```python
dados
```

```python
dados = pd.merge(
    left= dados,
    right= internet_df,
    left_index=True,
    right_index= True
)
dados
```

```python
dados = pd.merge(
    left= dados,
    right= account_df,
    left_index=True,
    right_index= True
)
dados
```

### Como as colunas customerID e Churn estão no DF principal e são de index 0 e 1 respectivamente,
### usaremos o insert para colocá-las no seu devido lugar da tabela

```python
dados.insert(loc=0,column='customerID',value=df.customerID)
```

```python
dados
```

```python
dados.insert(loc=1,column='Churn',value=df.Churn)
```

### Todos as linhas com Charges.Total sem dados nenhum são as mesma que tem a coluna tenure = 0,
### o que quer dizer que não tem nenhum mês de contrato

```python
# Fazendo testes para ver se os dados de Charges.Total é igual a tenure*Charges.Monthly
```

```python
# nesta análise conseguimos ver melhor
# Charges.Monthly é o gasto mensal do cliente e o Charges.Total é o gasto total do cliente 
# no tempo de empresa que ele tem (tenure)
dados[['tenure','Charges.Monthly','Charges.Total']]
```

```python
# Charges.Total é uma multiplicação de charges.monthly com tenure
teste = dados.tenure*dados['Charges.Monthly']
teste
```

```python
dados[teste == dados['Charges.Total']]
```

```python
# Os dados não são iguais mas são bem parecidos
# deve ter alguma relação com juros por atraso de pagamento, adicionais, troca de plano, etc.
```

```python
# analisando os dados onde tenure é zero e Charges.Total não existem.
dados[dados['Charges.Total'] == ' '][['tenure','Charges.Monthly','Charges.Total']]
```

Podemos levar em consideração que os dados de Charge.Total que estão faltando são os dados de Charge.Monthly
que por algum motivo não foram contabilizados ainda

```python
# pegando os índices do Charges.Total iguais a zero
idx = dados[dados['Charges.Total'] == ' '][['tenure','Charges.Monthly','Charges.Total']].index
idx
```

```python
# Igualando os gasto mensal ao gasto total, apenas dos índices que são tenure e Charges.Total iguais 0
dados.loc[idx,'Charges.Total']= dados.loc[idx,'Charges.Monthly']
```

```python
# Confirmando se deu tudo certo
dados[dados['Charges.Total'] == ' '][['tenure','Charges.Monthly','Charges.Total']]
```

```python
# Está OK!
dados.query('tenure == 0')[['tenure','Charges.Monthly','Charges.Total']]
```

```python
# Analisando as informações novamente.
dados.info()
```

A coluna Charges.Total está com object, sendo que ela é um valor em dinheiro e deve ser float.

```python
dados['Charges.Total'] = dados['Charges.Total'].astype('float')
```
