### Normalizando colunas
Como as colunas estÃ£o com os dados em formato json, precisamos normalizar para extrair os dados

```python
customer_df = pd.json_normalize(df.customer)
customer_df
```

```python
phone_df = pd.json_normalize(df.phone)
phone_df
```

```python
internet_df = pd.json_normalize(df.internet)
internet_df
```

```python
account_df = pd.json_normalize(df.account)
account_df
```

```python
df
```

### Unindo DataFrames gerados

Os DF gerados apÃ³s a normalizaÃ§Ã£o das colunas precisam ser unidos para que possamos 
dar continuidade no processo de anÃ¡lise dos dados. 
Vamos unir pelo Ã­ndice e de forma lateral, jÃ¡ que os dados tem os mesmos Ã­ndices

```python
dados = pd.merge(
    left= customer_df,
    right= phone_df,
    left_index=True,
    right_index= True
)
```

```python
dados
```

```python
dados = pd.merge(
    left= dados,
    right= internet_df,
    left_index=True,
    right_index= True
)
dados
```

```python
dados = pd.merge(
    left= dados,
    right= account_df,
    left_index=True,
    right_index= True
)
dados
```

### Como as colunas customerID e Churn estÃ£o no DF principal e sÃ£o de index 0 e 1 respectivamente,
### usaremos o insert para colocÃ¡-las no seu devido lugar da tabela

```python
dados.insert(loc=0,column='customerID',value=df.customerID)
```

```python
dados
```

```python
dados.insert(loc=1,column='Churn',value=df.Churn)
```

### Todos as linhas com Charges.Total sem dados nenhum sÃ£o as mesma que tem a coluna tenure = 0,
### o que quer dizer que nÃ£o tem nenhum mÃªs de contrato

```python
# Fazendo testes para ver se os dados de Charges.Total Ã© igual a tenure*Charges.Monthly
```

```python
# nesta anÃ¡lise conseguimos ver melhor
# Charges.Monthly Ã© o gasto mensal do cliente e o Charges.Total Ã© o gasto total do cliente 
# no tempo de empresa que ele tem (tenure)
dados[['tenure','Charges.Monthly','Charges.Total']]
```

```python
# Charges.Total Ã© uma multiplicaÃ§Ã£o de charges.monthly com tenure
teste = dados.tenure*dados['Charges.Monthly']
teste
```

```python
dados[teste == dados['Charges.Total']]
```

```python
# Os dados nÃ£o sÃ£o iguais mas sÃ£o bem parecidos
# deve ter alguma relaÃ§Ã£o com juros por atraso de pagamento, adicionais, troca de plano, etc.
```

```python
# analisando os dados onde tenure Ã© zero e Charges.Total nÃ£o existem.
dados[dados['Charges.Total'] == ' '][['tenure','Charges.Monthly','Charges.Total']]
```

Podemos levar em consideraÃ§Ã£o que os dados de Charge.Total que estÃ£o faltando sÃ£o os dados de Charge.Monthly
que por algum motivo nÃ£o foram contabilizados ainda

```python
# pegando os Ã­ndices do Charges.Total iguais a zero
idx = dados[dados['Charges.Total'] == ' '][['tenure','Charges.Monthly','Charges.Total']].index
idx
```

```python
# Igualando os gasto mensal ao gasto total, apenas dos Ã­ndices que sÃ£o tenure e Charges.Total iguais 0
dados.loc[idx,'Charges.Total']= dados.loc[idx,'Charges.Monthly']
```

```python
# Confirmando se deu tudo certo
dados[dados['Charges.Total'] == ' '][['tenure','Charges.Monthly','Charges.Total']]
```

```python
# EstÃ¡ OK!
dados.query('tenure == 0')[['tenure','Charges.Monthly','Charges.Total']]
```
